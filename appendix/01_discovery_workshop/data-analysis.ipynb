# Analysis of Discovery Workshop responses

## Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style("white")
```

## Dietary Factors

```python
# Data
labels = [
    "Energy content", "Macronutrients", "Micronutrients", "Secondary plant compounds", 
    "Dietary fibers", "Food groups", "Allergens", "Additives", 
    "Dietary supplements", "Meal times", 
    "Degree of food processing", "Sustainability (e.g., emissions, seasonality)", 
    "Specific scoring systems (e.g., Nutri-Score)"
]
values = [4, 7, 4, 2, 7, 6, 3, 3, 4, 4, 7, 3, 1]

# Create a DataFrame and remove rows with None values in the "Value" column
data = pd.DataFrame({"Category": labels, "Value": values}).dropna(subset=["Value"])

# Sort data in descending order by "Value"
data = data.sort_values(by="Value", ascending=False)

# Update plot parameters
plt.rcParams.update({
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'legend.title_fontsize': 16
})

# Plot with a dark gray color
plt.figure(figsize=(10, 8))
ax = sns.barplot(data=data, y="Category", x="Value", color="#a9a9a9")

# Set x-axis limit from 0 to 7
plt.xlim(0, 7)

# Add values to each bar with larger font size
for index, value in enumerate(data["Value"]):
    ax.text(value + 0.1, index, f"{int(value)}", color='black', va="center", fontsize=14)

plt.xlabel("Number of Participants Who Selected Dietary Factors as Important")
plt.ylabel("Dietary Factors")
plt.title("Importance of Dietary Factors for Food Purchase Data (Sorted Descending)", fontsize=16)

# Automatically adjust the layout for better readability
plt.tight_layout()

plt.show()
```

## Food Product Information

```python
# New data with line breaks for long labels
labels = [
    "Product nutrient content", 
    "Energy content of a purchase,\nper macronutrient, or food category", 
    "Ingredient list, additives,\nand allergens, etc.", 
    "Intake recommendations", 
    "Nutri-Score of individual products\nor average within a food category", 
    "Proportion of processed products", 
    "Quantity of purchased products\nand frequency of purchase", 
    "Ranking of most frequently\npurchased products", 
    "Ranking of products with the highest\nenergy content, saturated fat content, etc.", 
    "Ratios such as the proportion of\nanimal to plant protein, etc.", 
    "Spending per purchase", 
    "Adherence to general dietary recommendations\n(general guidelines, therapeutic guidelines)"
]
values = [5, 4, 1, 1, 0, 4, 2, 4, 4, 5, 2, 3]

# Create a DataFrame
data = pd.DataFrame({"Category": labels, "Value": values})

# Sort data in descending order by "Value"
data = data.sort_values(by="Value", ascending=False)

# Update plot parameters
plt.rcParams.update({
    'axes.labelsize': 16,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 14,
    'legend.title_fontsize': 16
})

# Plot with a darker gray color
plt.figure(figsize=(10, 8))
ax = sns.barplot(data=data, y="Category", x="Value", color="#a9a9a9")

# Set x-axis limit from 0 to 7 to match previous style
plt.xlim(0, 7)

# Add values to each bar with larger font size
for index, value in enumerate(data["Value"]):
    ax.text(value + 0.1, index, f"{int(value)}", color='black', va="center", fontsize=14)

plt.xlabel("Number of Participants Who Selected Food Product Information as Important")
plt.ylabel("Food Product Information")
plt.title("Importance of Food Product Information for Food Purchase Data (Sorted Descending)", fontsize=16)

# Automatically adjust the layout for better readability
plt.tight_layout()

plt.show()
```
